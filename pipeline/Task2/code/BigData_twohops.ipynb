```python
import pandas as pd
import numpy as np
```

```python
df_emt = pd.read_csv('emt_trxns.csv')
```

```python
df_emt.info()
```

```python
df_wire = pd.read_csv('wire_trxns.csv')
```

```python
df_wire.info()
```

```python
df_combined = pd.concat([df_emt, df_wire], ignore_index=True)
df_combined.info()
```

```python
df_buyer = pd.read_csv('31_seller.csv')
```

```python
df_buyer.info()
```

```python
buyer_ids = df_buyer['id'].tolist()
```

```python
#buyer_ids
```

```python
matches_upward = df_combined[df_combined['id receiver'].isin(buyer_ids)]
matches_upward['emt value'].fillna(0, inplace=True)
matches_upward['wire value'].fillna(0, inplace=True)
matches_upward['total amount'] = matches_upward['emt value'] + matches_upward['wire value']
matches_upward = matches_upward.drop(columns=['trxn_id', 'country sender', 'country receiver', 'emt value', 'wire value'])
matches_upward = matches_upward.groupby(['id sender', 'id receiver', 'name sender', 'name receiver'], as_index=False).agg({'total amount': 'sum'})
matches_upward['label_first_upward'] = '1'
matches_upward
```

```python
matches_upward.to_csv('first_upward_transactions.csv', index=False)
```

```python
unique_senders_first_upward = matches_upward['id sender'].unique()
```

```python
is_in_buyer_ids = np.isin(unique_senders_first_upward, buyer_ids, invert=True)
unique_senders_first_upward = unique_senders_first_upward[is_in_buyer_ids]
```

```python
matches_second_upward = df_combined[df_combined['id receiver'].isin(unique_senders_first_upward)]
matches_second_upward['emt value'].fillna(0, inplace=True)
matches_second_upward['wire value'].fillna(0, inplace=True)
matches_second_upward['total amount'] = matches_second_upward['emt value'] + matches_second_upward['wire value']
matches_second_upward = matches_second_upward.drop(columns=['trxn_id', 'country sender', 'country receiver', 'emt value', 'wire value'])
matches_second_upward = matches_second_upward.groupby(['id sender', 'id receiver', 'name sender', 'name receiver'], as_index=False).agg({'total amount': 'sum'})
matches_second_upward['label_second_upward'] = '1'
matches_second_upward
```

```python
#matches_second_upward['id sender'].isin(unique_senders_first_upward)
```

```python
matches_second_upward.to_csv('second_upward_transactions.csv', index=False)
```

```python
matches_downward = df_combined[df_combined['id sender'].isin(buyer_ids)]
unique_receiver_first_downward = matches_downward['id receiver'].unique()
matches_downward['emt value'].fillna(0, inplace=True)
matches_downward['wire value'].fillna(0, inplace=True)
matches_downward['total amount'] = matches_downward['emt value'] + matches_downward['wire value']
matches_downward = matches_downward.drop(columns=['trxn_id', 'country sender', 'country receiver', 'emt value', 'wire value'])
matches_downward = matches_downward.groupby(['id sender', 'id receiver', 'name sender', 'name receiver'], as_index=False).agg({'total amount': 'sum'})
matches_downward['label_first_downward'] = '1'
matches_downward
```

```python
matches_downward.to_csv('first_downward_transactions.csv', index=False)
```

```python
unique_receiver_first_downward = matches_downward['id receiver'].unique()
```

```python
is_in_buyer_ids = np.isin(unique_receiver_first_downward, buyer_ids, invert=True)
unique_receiver_first_downward = unique_receiver_first_downward[is_in_buyer_ids]
```

```python
matches_second_downward = df_combined[df_combined['id sender'].isin(unique_receiver_first_downward)]
matches_second_downward['emt value'].fillna(0, inplace=True)
matches_second_downward['wire value'].fillna(0, inplace=True)
matches_second_downward['total amount'] = matches_second_downward['emt value'] + matches_second_downward['wire value']
matches_second_downward = matches_second_downward.drop(columns=['trxn_id', 'country sender', 'country receiver', 'emt value', 'wire value'])
matches_second_downward = matches_second_downward.groupby(['id sender', 'id receiver', 'name sender', 'name receiver'], as_index=False).agg({'total amount': 'sum'})
matches_second_downward['label_second_downward'] = 1
matches_second_downward
```

```python
matches_second_downward.to_csv('second_downward_transactions.csv', index=False)
```

```python
df1 = pd.read_csv('first_upward_transactions.csv')
df2 = pd.read_csv('second_upward_transactions.csv')
df3 = pd.read_csv('first_downward_transactions.csv')
df4 = pd.read_csv('second_downward_transactions.csv')
```

```python
dfs = [df1, df2, df3, df4]
```

```python
for df in dfs:
    for label in ['label_first_upward', 'label_second_upward', 'label_first_downward', 'label_second_downward']:
        if label not in df.columns:
            df[label] = '0'
```

```python
df_combined = pd.concat(dfs, ignore_index=True)
df_combined
```

```python
duplicates = df_combined.duplicated(subset=['id sender', 'id receiver','name sender', 'name receiver'], keep=False)
df_duplicates = df_combined[duplicates]
df_duplicates
```

```python
df_combined = df_combined.drop_duplicates(subset=['id sender', 'id receiver','name sender', 'name receiver'])
df_combined
```

```python
df_combined.to_csv('new_two_hops_trans.csv', index=False)
```
